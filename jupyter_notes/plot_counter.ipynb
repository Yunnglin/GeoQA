```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from raw_data_process import Word
from utils.json_io import read_json
from matplotlib.pyplot import MultipleLocator
```

```python

data_path = '../data/processed/data_no_graph/no_cut/redundant/length_count.json'
data = read_json(data_path)
print(data)

length = []
frequency = []
for l , f in data.items():
    length.append(l)
    frequency.append(f)
length = np.array(length,dtype=int)
frequency = np.array(frequency,dtype=int)
print(length)
print(frequency)

total_data = []
for l,f in data.items():
    total_data+=[l]*f
print(total_data)

total_data = np.array(total_data,dtype=int)
```

```python
#  length  frequency 
plt.hist(total_data,bins=200,cumulative=True,density=True)
plt.title("Question Length Cumulative Frequency")
plt.xlabel("question length")
plt.ylabel("frequency")
plt.xlim(10,400)
plt.ylim(0,1)
plt.savefig("../result/figures/ques_len_cum_freq.png")
plt.show()
```

```python
plt.hist(total_data,bins=100)
plt.grid()
plt.title("Question Length Histogram")
plt.xlabel("question length")
plt.ylabel("frequency")
plt.xlim(10,400)
plt.savefig("../result/figures/ques_len_freq.png")
plt.show()
```

```python
def cal_percentage(x,datas):
    count = 0
    for t in datas:
        if t<=x:
            count+=1
    return count/datas.size
cal_percentage(170,total_data)# cover 97%
```

```python
data_path = '../data/processed/data_no_graph/no_cut/redundant/word_count.json'
data = read_json(data_path)
word = np.array(list(data.values()),dtype=int)

plt.hist(word,bins=2000,log=False, cumulative=True,density=True)
plt.xlim(1,20)
x_major_locator = MultipleLocator(1)

ax=plt.gca()
#ax
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(MultipleLocator(0.1))
plt.title("Log_counts of Each Word Frequency")
plt.grid()
plt.xlabel("word frequency")
plt.ylabel("word count")
plt.savefig('../result/figures/count_word_cum_freq.png')
plt.show()
```
